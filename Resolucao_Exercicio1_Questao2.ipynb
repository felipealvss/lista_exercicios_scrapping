```python
import requests
from bs4 import BeautifulSoup
```

```python
link = 'https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops'

res = requests.get(link)

if res.status_code == 200:

    print(f"Sucesso: {res}")
    soup = BeautifulSoup(res.text, 'html.parser')
else:
    print(f"Erro de requisição: {res.status_code}")
```

```python
lista_itens = soup.find_all('div', class_='card-body')
```

```python
for i, item in enumerate(lista_itens):
    titulo = item.find('a', class_='title').text.strip()
    descricao = item.find('p', class_='description').text.strip()
    preco = item.find('h4', class_='float-end price card-title pull-right').text.strip()
    link_produto = item.find('a')['href']
    
    print(f'Produto No. {i + 1}:')
    print(f'Título: {titulo}')
    print(f'Descrição: {descricao}')
    print(f'Preço: {preco}')
    print(f'Link produto: {link_produto}')
    print()
    print('-----------------------------------------------------------------------------------------------')
    print()
```
